```python
import numpy as np
from setMassInVrep import getMassfromSW

f = open('mass.txt', 'w')
for i in range(8):
    ms, it, ct = getMassfromSW('arm'+str(i))
    f.write(str(ms))
    f.write('\n')
    f.write(str(it.reshape(9,1)))
    f.write('\n')
    f.write(str(ct.reshape(3,1)))
    f.write('\n')
    
f.close()
```
